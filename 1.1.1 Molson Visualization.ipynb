```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Import the Excel

## Create the Revenue Column based on the Intensity and Total Environmental Cost

```python
# Load the dataset
df = pd.read_csv("Sales_no_outliers_percentage.csv", index_col=0)
df = df.drop("Industry (Exiobase)", axis=1)
df.head()
```

## Filter for only Molson Company

```python
molson_df = df[df['Company Name'] == 'MOLSON COORS BEVERAGE CO']
# Group the data by "Company Name" and count unique years for each group
molson_df
```

# Boxplots


## On SDGs

```python
df_opinc_box = molson_df.melt(id_vars='GICS Sub-Industry', value_vars=molson_df.loc[:,"SDG 1.5":"SDG 15.5"])
df_opinc_box
```

```python
# Create the boxplot
plt.figure(figsize=(20, 8))
sns.boxplot(x='variable', y='value', data=df_opinc_box)
plt.xlabel('Metrics')
plt.ylabel('Percentage of Operating Income (%)')
plt.title('Boxplot of SDGs as Percentage of Op income')
plt.show()
```

```python
for column in molson_df.loc[:,"SDG 1.5":"SDG 15.5"]:  # Exclude the 'Year' column
    sns.lineplot(x='Year', y=column, data=molson_df, label=column)
sns.lineplot(x='Year', y="Environmental Intensity (Sales)", data=molson_df, label="Total Cost")

# Add labels and a legend
plt.xlabel('Year')
plt.ylabel('Value %')
plt.title('SDGs Changes over time Molson')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))  # Adjust legend position

# Show the plot
plt.show()
```
